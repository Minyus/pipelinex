```python
from pipelinex import log_time
from pipelinex import (
    mem_profile,
)  # Need to install memory_profiler for memory profiling
from pipelinex import nvml_profile  # Need to install pynvml for NVIDIA GPU profiling
from time import sleep
import logging

logging.basicConfig(level=logging.INFO)


@nvml_profile
@mem_profile
@log_time
def foo_func(i=1):
    sleep(
        0.5
    )  # Needed to avoid the bug reported at https://github.com/pythonprofilers/memory_profiler/issues/216
    return "a" * i


output = foo_func(100_000_000)
```
